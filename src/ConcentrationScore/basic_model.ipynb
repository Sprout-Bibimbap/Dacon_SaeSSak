```python
import tensorflow as tf
from transformers import AutoTokenizer, TFBertForNextSentencePrediction

# Load model and tokenizer
model = TFBertForNextSentencePrediction.from_pretrained('klue/bert-base', from_pt=True)
tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")

def predict_next_sentence(sentence_a, sentence_b):
    # Tokenize input
    inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='tf')

    # Get logits from the model
    outputs = model(inputs['input_ids'], token_type_ids=inputs['token_type_ids'])
    logits = outputs.logits

    # Apply softmax to get probabilities
    softmax = tf.keras.layers.Softmax()
    probs = softmax(logits)

    is_next_prob = probs[0][0].numpy()
    not_next_prob = probs[0][1].numpy()

    return is_next_prob, not_next_prob

# Define sentences
sentence_a_1 = "나는 이 인형이 싫어"
sentence_a_2 = "그래? 그거 귀엽지 않아?"

sentence_b_1 = "내일은 이모네 놀러갈거에요"
sentence_b_2 = "게임은 항상 재밌어"

# Get probabilities for next sentence predictions
next_a, not_next_a = predict_next_sentence(sentence_a_1, sentence_a_2)
next_b, not_next_b = predict_next_sentence(sentence_b_1, sentence_b_2)

# Print results
print(f'Next sentence A: {next_a}, Not next sentence A: {not_next_a}')
print(f'Next sentence B: {next_b}, Not next sentence B: {not_next_b}')
```
